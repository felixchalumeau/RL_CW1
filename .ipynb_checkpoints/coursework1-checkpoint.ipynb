```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = 0.75
gamma = 0.3
```

## Useful class & functions definition

```python
from rl_algorithms import *
```

### Define the grid & my personal parameters

```python
### Define the grid
print("Creating the Grid world, represented as:\n")
grid = GridWorld(p)
```

## Little warm-up with da uniform policy

```python
# Plot policy
Policy = np.zeros((grid.state_size, grid.action_size))
Policy = Policy + 0.25

# Policy evaluation
val, epochs = grid.policy_evaluation(Policy, gamma, 0.00001)

# Plot value function
grid.draw_value(val)

# Plot number of epochs
print("It took {} epochs".format(epochs))
```

## b. Dynamic Programming

```python
# Policy iteration algorithm
V_opt_dp, pol_opt_dp, epochs = grid.policy_iteration(gamma, 0.0001)

# Plot value function for policy iteration
grid.draw_value(V_opt_dp, save_link="dp_optimal_value_function.png")

# Plot policy for policy iteration
grid.draw_deterministic_policy(np.array([np.argmax(pol_opt_dp[row,:]) for row in range(grid.state_size)]), save_link="dp_optimal_policy.png")

# Plot number of epochs
print("It took {} epochs".format(epochs))
```

## c. Monte Carlo RL

```python
Q = np.zeros((grid.state_size, 4))
Q_2 = np.zeros((grid.state_size, 4)) + np.array([0, 1, 1, 0])
trace = grid.simulate_episode(Q, epsilon=0.05, verbose=False)

trace[0::3], trace[1::3], (len(trace) - 1)/3
```

```python
# Policy iteration algorithm
Q_opt, pol_opt, epochs, rewards, value_functions = grid.montecarlo_on_control(epsilon_init=0.05, decay_speed=0, decay_smooth=1, discount=gamma, learning_rate=0.005, threshold=0.0001, max_epoch=30000)

# Plot value function for policy iteration
#print("The value of the optimal policy computed using policy iteration is:\n\n {}".format(V_opt))
#print("\n\nIts graphical representation is:\n")
V_opt = np.max(Q_opt, axis=1)
grid.draw_value(V_opt, save_link="figures/mc_optimal_value_function.png")

# Plot policy for policy iteration
#print("\n\nThe optimal policy using policy iteration is:\n\n {}".format(pol_opt))
print("\n\nIts graphical representation is:\n")
grid.draw_deterministic_policy(np.array([np.argmax(pol_opt[row,:]) for row in range(grid.state_size)]), save_link="figures/mc_optimal_policy.png")

# Plot number of epochs
print("\nIt took {} epochs\n".format(epochs))
```

```python
np.sqrt(np.sum(np.square(value_functions - V_opt_dp)))
```

```python
estimation_error = np.sqrt(np.sum(np.square(value_functions - V_opt_dp), axis=1))
estimation_error
```

```python
plt.figure(figsize=(12, 5))
plt.plot([i+1 for i in range(0, len(estimation_error))], estimation_error)
plt.xlabel("Episode")
plt.ylabel("Estimation error")
plt.grid()
plt.savefig("figures/montecarlo_estimation_error.png")
plt.show()
```

```python
np.round(np.max(Q_opt, axis=1), 2), np.round(V_opt_dp, 2)
```

```python
# Policy evaluation
val, epochs = grid.policy_evaluation(pol_opt, gamma, 0.00001)

# Plot value function
grid.draw_value(val)

# Plot number of epochs
print("It took {} epochs".format(epochs))
```

```python
Q_opt, pol_opt, epochs, rewards = grid.montecarlo_on_control(epsilon_init=0.5, decay_speed=0, decay_smooth=1, discount=gamma, learning_rate=0.5, threshold=0.0001, max_epoch=300)
rewards_matrix = np.array([rewards])
for i in range(0, 400):
    Q_opt, pol_opt, epochs, rewards = grid.montecarlo_on_control(epsilon_init=0.5, decay_speed=0, decay_smooth=1, discount=gamma, learning_rate=0.5, threshold=0.0001, max_epoch=300)
    rewards_matrix = np.concatenate((rewards_matrix, np.array([rewards])))

mean_rewards = np.mean(rewards_matrix, axis=0)
std_rewards = np.std(rewards_matrix, axis=0)
plt.figure(figsize=(12, 5))
plt.plot([i+1 for i in range(0, len(mean_rewards))], mean_rewards)
plt.plot([i+1 for i in range(0, len(mean_rewards))], mean_rewards + std_rewards, color="red")
plt.plot([i+1 for i in range(0, len(mean_rewards))], mean_rewards - std_rewards, color="red")
plt.xlabel("Episode")
plt.ylabel("Reward of the episode")
plt.grid()
plt.savefig("figures/montecarlo_control_epsilon_05.png")
plt.show()
```

## d. Temporal Difference RL

```python
# Policy iteration algorithm
Q_opt, pol_opt, epochs, rewards, value_functions = grid.sarsa_on_td_control(epsilon_init=0.05, decay_speed=0, decay_smooth=1, discount=gamma, learning_rate=0.005, threshold=0.0001, max_epoch=10000)

# Plot value function for policy iteration
#print("The value of the optimal policy computed using policy iteration is:\n\n {}".format(V_opt))
#print("\n\nIts graphical representation is:\n")
V_opt = np.max(Q_opt, axis=1)
grid.draw_value(V_opt, save_link="figures/sarsa_optimal_value_function.png")

# Plot policy for policy iteration
#print("\n\nThe optimal policy using policy iteration is:\n\n {}".format(pol_opt))
print("\n\nIts graphical representation is:\n")
grid.draw_deterministic_policy(np.array([np.argmax(pol_opt[row,:]) for row in range(grid.state_size)]), save_link="figures/sarsa_optimal_policy.png")

# Plot number of epochs
print("\nIt took {} epochs\n".format(epochs))
```

```python
estimation_error = np.sqrt(np.sum(np.square(value_functions - V_opt_dp), axis=1))
plt.figure(figsize=(12, 5))
plt.plot([i+1 for i in range(0, len(estimation_error))], estimation_error)
plt.xlabel("Episode")
plt.ylabel("Estimation error")
plt.grid()
plt.savefig("figures/sarsa_estimation_error.png")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(mean_rewards, estimation_error[1:])
plt.xlabel("Reward")
plt.ylabel("Estimation error")
plt.grid()
plt.savefig("figures/sarsa_estimation_error_vs_rewards.png")
plt.show()
```
